```python
import os, pickle
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
```

```python
model_name = 'Luo6ChunksNoPad2LstmNormalisedAudioSeed2'
file_name = 'pred_target_labels_log.pkl'
root_path = './results'
file_path = os.path.join(root_path,model_name,file_name)
label_data = pickle.load(open(file_path, 'rb'))
```

```python
print('number of epochs in label_data: ', len(label_data))
print('number of sets in epoch (train and evaluations) :', len(label_data[0]))
print('number of examples in training set: ', len(label_data[0][0]))
print('number of examples in eval set: ', len(label_data[0][1]))
```

```python
# look at the data for the eval set in the last epoch
examined_epoch = len(label_data)-1


predictions = []
targets = []
for eval_data in label_data[examined_epoch][0]:
    predictions.append(eval_data[0])
    targets.append(eval_data[1])

names = ['blt','lip','str','fry','vib','bre']
conf_mat = confusion_matrix(targets, predictions)
fig, ax = plt.subplots(figsize=(20, 10))
im = ax.imshow(conf_mat)
ax.set_xticks(np.arange(6))
ax.set_yticks(np.arange(6))
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
for i in range(6):
    for j in range(6):
        text = ax.text(j, i, conf_mat[i, j],
                       ha="center", va="center", color="w")
        
predictions = []
targets = []
for eval_data in label_data[examined_epoch][1]:
    predictions.append(eval_data[0])
    targets.append(eval_data[1])

names = ['blt','lip','str','fry','vib','bre']
conf_mat = confusion_matrix(targets, predictions)
fig, ax = plt.subplots(figsize=(20, 10))
im = ax.imshow(conf_mat)
ax.set_xticks(np.arange(6))
ax.set_yticks(np.arange(6))
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
for i in range(6):
    for j in range(6):
        text = ax.text(j, i, conf_mat[i, j],
                       ha="center", va="center", color="w")
```
