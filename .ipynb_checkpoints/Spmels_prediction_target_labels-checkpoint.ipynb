```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data_batch = np.load('./saved_examples/x_data_e1_b0.npy')
print(x_data_batch.shape)
y_data_batch = np.load('./saved_examples/y_data_e1_b0.npy')
print(y_data_batch.shape)
```

```python
chunk_num = 6
singer_id_batch = np.load('./saved_examples/singer_id_e1_b0.npy')
y_data_batch = np.load('./saved_examples/y_data_e1_b0.npy')
x_data_batch = np.load('./saved_examples/x_data_e1_b0.npy')
num_examples = x_data_batch.shape[0]//chunk_num
techs = ['belt','lip_trill','straight','vocal_fry','vibrato','breathy']

for i in range(num_examples):
    offset = i * chunk_num
    example_data = x_data_batch[offset: offset+chunk_num]
    for j in range(example_data.shape[0]):
        if j==0:
            catted_data = example_data[j]
        else:
            catted_data = np.concatenate((catted_data, example_data[j]))
    print('Example ', i, 'Technique', techs[y_data_batch[i]], 'Singer', singer_id_batch[i])
    spmel = np.rot90(catted_data)
    plt.imshow(spmel, cmap='gray')
    plt.show()
```
